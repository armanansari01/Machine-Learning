# Cleaning Data:
 * URLs
 * Hashtags
 * Mentions
 * Special characters
 * punctuation

```python
import re
def clean_resume(txt):
    #removing urls
    cleanTxt = re.sub('http\S+\s',' ',txt)
    cleanTxt = re.sub('RT|cc',' ',cleanTxt)

    #removing hashtags
    cleanTxt = re.sub('#\S+\s',' ',cleanTxt)
    #removing @ or gmails
    cleanTxt = re.sub('@\S+\s',' ',cleanTxt)
    #removing special characters
    cleanTxt = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""),' ',cleanTxt)
    cleanTxt = re.sub(r'[^\x00-\x7f]',' ',cleanTxt)
    cleanTxt = re.sub('\s+',' ',cleanTxt)

    return cleanTxt
```

```python
clean_resume("Hi this is ##hwnlp my website link is this https://helloworld and access it using @gmail.com of my blog post #hw .")
```

```python
sample = df['Resume'][0]
sample
```

```python
clean_resume(sample)
```

```python
df['Cleaned_resume'] = df['Resume'].apply(lambda x: clean_resume(x)) 
```

```python
df['Cleaned_resume']
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(df['Category'])
df['encoded_category'] = le.transform(df['Category'])
```

```python
df[['Category','encoded_category']].value_counts()
```

## Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
```

```python
tfidf.fit(df['Cleaned_resume'])
vectorized_resume = tfidf.transform(df['Cleaned_resume'])
```

```python
vectorized_resume
```

# Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(vectorized_resume, df['encoded_category'], test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

# Model Training

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score
```

```python
clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
print(y_pred)
```

```python
print(accuracy_score(y_test,y_pred))
```
